```python
###################### preprocessing #################
from sklearn.preprocessing import *
from sklearn.model_selection import *
from sklearn.cluster import KMeans,DBSCAN,AffinityPropagation
from sklearn.metrics import *
from sklearn.metrics.cluster import contingency_matrix
from sklearn.mixture import *



marketing2=marketing.copy()
def Imputer(data):
    data=data.drop(['ID','Z_CostContact','Z_Revenue'],axis=1)
    data.dropna(axis=0,inplace=True)
    return data 


def Encodage(data):
    codage={'Single':0,
          'Together':1,
          'Married':2,
          'Divorced':3,
          'Widow':4,
          'Alone':5,
          'Absurd':6,
          'YOLO':7,
          'Graduation':0,
          'PhD':1,
          'Master':2,
          'Basic':3,
          '2n Cycle':4
            }
    for col in data.select_dtypes('object'):
        data[col]=data[col].map(codage)
    
    return data 

def Modele(data):
    data=Imputer(data)
    data=Encodage(data)
    
    x=data.drop('Dt_Customer',axis=1)
    
    return x

dataset=Modele(marketing)
print("Shape Data=>",dataset.shape)

def Eva_modele(model):
    model.fit(dataset)
    print("Prediction",model.predict(dataset))
    #print("labels",model.labels_)
    print("score_1",silhouette_score(dataset,model.predict(dataset), metric='euclidean'))
    #print("score_2",calinski_harabasz_score(xdata,model.labels_))
    #print("score_3",davies_bouldin_score(xdata,model.labels_))
    #print("matrice contingence",contingency_matrix(xdata,model.labels_))
    #print("Centre Cluster",model.cluster_centers_)
    #plt.scatter(dataset.iloc[:,7].values,dataset.iloc[:,8].values,c=model.predict(dataset))
    #plt.show()


#Eva_modele(dataset,GaussianMixture(n_components=3,random_state=0))
#Eva_modele(KMeans(n_clusters=3))

#elbow methode 
cout=[]
for i in range(1,15):
    mod=KMeans(n_clusters=i)
    mod.fit(dataset)
    cout.append(mod.inertia_)

print(cout)
k=np.arange(1,15)
fig1=plt.figure()
plt.plot(k,cout)

```

```python
from sklearn.pipeline import *
from sklearn.feature_selection import *

print(round(marketing.var()))

KM=make_pipeline(VarianceThreshold(threshold=4),KMeans(n_clusters=3))
GM=make_pipeline(VarianceThreshold(threshold=4),GaussianMixture(n_components=3,random_state=0))

liste_modele={
    'KMeans':KM,
    'Melange gaussian':GM
}

for nom,model in liste_modele.items():
    print("")
    print(nom)
    Eva_modele(model)


```
